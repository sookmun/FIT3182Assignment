```python
from pprint import pprint
from datetime import datetime
import csv
import json
climateArray=[]
with open('climate_streaming.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    hotspot_pointer=-1
    for row in csv_reader:
        if line_count!=0:
            temp={"latitude" :0,"longitude":0,"air_temperature_celcius":0,"relative_humidity":0,
                  "windspeed_knots":0,"max_wind_speed":0,"precipitation":0,"precip_flag":"","GHI_w/m2":0,"date:":""}
            temp["latitude"]=float(row[0])
            temp["longitude"]=float(row[1])
            temp["air_temperature_celcius"]=int(row[2])
            temp["relative_humidity"]=float(row[3])
            temp["windspeed_knots"]=float(row[4])
            temp["max_wind_speed"]=float(row[5])
            temp["precipitation"]=float(row[6][:-1])
            if row[6][-1] in ["A","B","C","D","E","F","G","H","I"]:
                temp["precip_flag"]=row[6][-1]
            else:
                temp["precip_flag"]="Z"
            temp["GHI_w/m2"]=int(row[7])
            climateArray.append(temp) 
            
        line_count+=1
print(type(climateArray[0]))
new_string = json.dumps(climateArray[0], sort_keys=True, indent=3)
new_string
print(type(new_string))
```

```python
# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt

 
def publish_message(producer_instance, topic_name, key,data):
    try:
        value_str = json.dumps(data, sort_keys=True, indent=3)
        value_bytes=bytes(value_str, 'utf-8')
        key_bytes=bytes(key, 'utf-8')
        #value_bytes = str(data).encode(encoding='UTF-8',errors='strict')
        #key_bytes = key.encode(encoding='UTF-8',errors='strict')
        print("HIHIHI")
        producer_instance.send(topic_name,  key=key_bytes, value=value_bytes)
        print('Message published successfully. ' + value_str)
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['127.0.0.1:9092'],
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'partB'
    print('Publishing records..')
    producer = connect_kafka_producer()
    date_time_str='31/12/2018' #latest day
    #date = datetime.strptime(date_time_str, '%d/%m/%Y')
    for e in range(len(climateArray)):
        random_index=random.randrange(0,len(climateArray))
        data=climateArray[random_index]
        print(type(data))
        #date =date+ datetime.timedelta(days=1)
        #data["date"]=date.date()
        #print(climateArray[random_index]) 
        publish_message(producer, topic, "producer1", data)
        sleep(10)
```
