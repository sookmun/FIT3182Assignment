```python
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt
from pprint import pprint
import datetime
import csv
import json

 
def publish_message(producer_instance, topic_name, key,data):
    try:
        #encoding the key and value into bytes
        value_str = json.dumps(data, sort_keys=True, indent=3)
        value_bytes=bytes(value_str, 'utf-8')
        key_bytes=bytes(key, 'utf-8')

        producer_instance.send(topic_name,  key=key_bytes, value=value_bytes)
        print('Message published successfully. ' + value_str)
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['127.0.0.1:9092'],
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer

def get_data():
    #getting data from the aqua csv and storing into an array
    aquaArray=[]
    with open('hotspot_AQUA_streaming.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        hotspot_pointer=-1
        for row in csv_reader:
            if line_count!=0:
                temp={'latitude':0, 'longitude':0, 'confidence':0,
                      'surface_temperature_celcius':0,'date':"",'time':""}
                temp["latitude"]=float(row[0])
                temp["longitude"]=float(row[1])
                temp["confidence"]=int(row[2])
                temp["surface_temperature_celcius"]=int(row[3])  
                aquaArray.append(temp)
            line_count+=1
        return aquaArray
    
if __name__ == '__main__':
   
    topic = 'partB'
    print('Publishing records..')
    producer = connect_kafka_producer()
    
    date_time_str= str(datetime.datetime.now()) #the time set is now
    date = datetime.datetime.strptime(date_time_str, "%Y-%m-%d %H:%M:%S.%f")
    
    aquaArray=get_data()
    for e in range(len(aquaArray)):
        random_index=random.randrange(0,len(aquaArray))
        data=aquaArray[random_index]
        
        date =date+ datetime.timedelta(hours=4.8)#adding this because if 10 sec =24, then 2 sec is 4.8
        data["time"]=date.strftime("%H:%M:%S")#formating the time
        
        publish_message(producer, topic, "producer2",data) #specify the producer
        sleep(2)
```
