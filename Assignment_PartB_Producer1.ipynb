```python
# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime
from pprint import pprint
import datetime
import csv
import json
 
def publish_message(producer_instance, topic_name, key,data):
    try:
        #encoding the key and value into bytes
        value_str = json.dumps(data, sort_keys=True, indent=3)
        value_bytes=bytes(value_str, 'utf-8')
        key_bytes=bytes(key, 'utf-8')
        producer_instance.send(topic_name,  key=key_bytes, value=value_bytes)
        print('Message published successfully. ' + value_str)
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['127.0.0.1:9092'],
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
def get_data():
    climateArray=[]
    with open('climate_streaming.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        hotspot_pointer=-1
        for row in csv_reader:
            if line_count!=0:
                temp={"latitude" :0,"longitude":0,"air_temperature_celcius":0,"relative_humidity":0,
                      "windspeed_knots":0,"max_wind_speed":0,"precipitation":0,"precip_flag":"","GHI_w/m2":0,"hotspots":[],
                      "date":""}
                temp["latitude"]=float(row[0])
                temp["longitude"]=float(row[1])
                temp["air_temperature_celcius"]=int(row[2])
                temp["relative_humidity"]=float(row[3])
                temp["windspeed_knots"]=float(row[4])
                temp["max_wind_speed"]=float(row[5])
                temp["precipitation"]=float(row[6][:-1])
                if row[6][-1] in ["A","B","C","D","E","F","G","H","I"]:
                    temp["precip_flag"]=row[6][-1]
                else:
                    temp["precip_flag"]="Z"
                temp["GHI_w/m2"]=int(row[7])
                climateArray.append(temp) 

            line_count+=1
        return climateArray
    
if __name__ == '__main__':
   
    topic = 'partB'
    print('Publishing records..')
    producer = connect_kafka_producer()
    date_time_str='31/12/2018' #latest day
    date = datetime.datetime.strptime(date_time_str, '%d/%m/%Y') #format the day value
    climateArray=get_data()
    for e in range(len(climateArray)):
        random_index=random.randrange(0,len(climateArray)) #get random data
        data=climateArray[random_index]
        date =date+ datetime.timedelta(days=1) #imcrement the date
        data["date"]=str(date.date())  #store the date as string
        publish_message(producer, topic, "producer1", data) #specify the key
        sleep(10)
```
